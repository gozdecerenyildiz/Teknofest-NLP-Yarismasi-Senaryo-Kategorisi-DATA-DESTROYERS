```python
import pandas as pd

df = pd.read_csv("df.csv")
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Gerekli NLTK paketlerini indir
nltk.download('punkt')
nltk.download('stopwords')

# Türkçe durak kelimeleri al
stop_words = set(stopwords.words('turkish'))

def clean_text_nltk(text):
    if isinstance(text, float):
        text = ""
    text = text.lower()
    text = re.sub(r'\s+', ' ', text)  # Çoklu boşlukları tek boşluğa indir
    text = re.sub(r'[^\w\s]', '', text)  # Noktalama işaretlerini kaldır
    words = word_tokenize(text)  # Tokenize et
    words = [word for word in words if word not in stop_words]  # Durak kelimeleri çıkar
    words = [word for word in words if word.isalpha()]  # Sadece alfabetik karakterleri tut
    return ' '.join(words)

df['cleaned_comment_nltk'] = df['yorum'].apply(clean_text_nltk)
print(df)
```

```python
# Bir veya daha fazla kolonu silin
kolonlar = ['product_name', 'Kategori']
df2 = df.drop(columns=kolonlar)
df2
```

```python
import stanza
from transformers import pipeline

# Stanza modeli indir ve yükle
stanza.download('tr')
nlp = stanza.Pipeline('tr')

def clean_text_stanza(text):
    doc = nlp(text)
    words = []
    for sentence in doc.sentences:
        for word in sentence.words:
            if word.text not in stop_words and word.upos != 'PUNCT':
                if word.lemma is not None:  # None değerlerini kontrol et
                    words.append(word.lemma)  # Lemma (kök) formunu kullan
    return ' '.join(words)

df2['cleaned_comment_stanza'] = df2['cleaned_comment_nltk'].apply(clean_text_stanza)
print(df2)
```

```python
sentiment_analyzer = pipeline('sentiment-analysis', model='dbmdz/bert-base-turkish-cased', tokenizer='dbmdz/bert-base-turkish-cased')

# Entity Extraction
def extract_entities(text):
    doc = nlp(text)
    entities = []
    for sentence in doc.sentences:
        for ent in sentence.ents:
            entities.append({
                'text': ent.text,
                'type': ent.type
            })
    return entities

# Sentiment Analysis
def analyze_sentiment(text):
    result = sentiment_analyzer(text)[0]
    return result['label'], result['score']

# Birleştirilmiş İşleme Fonksiyonu
def process_comment(comment):
    entities = extract_entities(comment)
    sentiment_label, sentiment_score = analyze_sentiment(comment)
    return {
        'entities': entities,
        'sentiment_label': sentiment_label,
        'sentiment_score': sentiment_score
    }
```
