```python
import pandas as pd

# Örnek veri çerçevesi
df = pd.read_csv('text_data.csv')

# Etiketlerin karşılıklarını tanımla
label_mapping = {
    0: "Pozitif",
    1: "Negatif",
    2: "Nötr"
}

# Yeni bir sütun oluştur ve label sütununu metin etiketlerine dönüştür
df['sentiment'] = df['label'].map(label_mapping)

# Sonuçları kontrol et
print(df.head())

# Düzenlenmiş veri çerçevesini kaydet
df.to_csv('sentiment_data.csv', index=False)
```

```python
df.dropna(inplace=True)
```

```python
positive_reviews = ' '.join(df[df.sentiment=='Pozitif'].temizlenmiş_yorum)
negative_reviews = ' '.join(df[df.sentiment=='Negatif'].temizlenmiş_yorum)
neutral_reviews = ' '.join(df[df.sentiment=='Nötr'].temizlenmiş_yorum)
```

```python
X = df.temizlenmiş_yorum
y = df.label
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## CountVectorizer

```python
# Sıkıştırılmış veri temsillerini kullanarak işlemi gerçekleştirin
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

cv1 = CountVectorizer(max_df=0.8, max_features=10000)
X_train_cv1 = cv1.fit_transform(X_train)
X_test_cv1  = cv1.transform(X_test)

# Veri çerçevesini sadece ilk 10 satır için oluşturun
df_train_cv1 = pd.DataFrame(X_train_cv1[:10].toarray(), columns=cv1.get_feature_names_out())
print(df_train_cv1.head())
```

```python
# 2'li N-Gram yöntemini kullanarak tokenize etme
cv2 = CountVectorizer(ngram_range=(1, 2), binary=True, max_df=0.8, max_features=2500)  # En çok görülen 2500 tokeni dahil et

# Train ve Test veri setlerine CountVectorizer uygulama
X_train_cv2 = cv2.fit_transform(X_train)
X_test_cv2  = cv2.transform(X_test)

# DataFrame'e dönüştürme
df_train_cv2 = pd.DataFrame(X_train_cv2.toarray(), columns=cv2.get_feature_names_out())
print(df_train_cv2.head())
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
# Kelimelere ayırarak oluşturulan DataFrame
lr.fit(X_train_cv1, y_train)
y_pred_cv1 = lr.predict(X_test_cv1)
```

```python
# 2'li N-Gram yöntemi kullanılarak oluşturulan DataFrame
lr.fit(X_train_cv2, y_train)
y_pred_cv2 = lr.predict(X_test_cv2)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def conf_matrix(actual, predicted):
    cm = confusion_matrix(actual, predicted, labels=[0, 1, 2])
    plt.figure(figsize=[8,8])
    sns.heatmap(cm, xticklabels=['Predicted Positive', 'Predicted Negative', 'Predicted Neutral'], 
                yticklabels=['Actual Positive', 'Actual Negative', 'Actual Neutral'], annot=True,
                fmt='d', annot_kws={'fontsize':35}, cmap="YlGnBu", square=True)
    
    accuracy = accuracy_score(actual, predicted)
    
    precision_pos = precision_score(actual, predicted, labels=[0], average='macro', zero_division=1)
    precision_neg = precision_score(actual, predicted, labels=[1], average='macro', zero_division=1)
    precision_neu = precision_score(actual, predicted, labels=[2], average='macro', zero_division=1)
    
    recall_pos = recall_score(actual, predicted, labels=[0], average='macro', zero_division=1)
    recall_neg = recall_score(actual, predicted, labels=[1], average='macro', zero_division=1)
    recall_neu = recall_score(actual, predicted, labels=[2], average='macro', zero_division=1)
    
    f1_pos = f1_score(actual, predicted, labels=[0], average='macro', zero_division=1)
    f1_neg = f1_score(actual, predicted, labels=[1], average='macro', zero_division=1)
    f1_neu = f1_score(actual, predicted, labels=[2], average='macro', zero_division=1)
    

    cm_results = pd.DataFrame({
        'Metric': ['Accuracy', 'Precision Positive', 'Precision Negative', 'Precision Neutral',
                   'Recall Positive', 'Recall Negative', 'Recall Neutral', 
                   'F1 Score Positive', 'F1 Score Negative', 'F1 Score Neutral'],
        'Score': [accuracy, precision_pos, precision_neg, precision_neu,
                  recall_pos, recall_neg, recall_neu,
                  f1_pos, f1_neg, f1_neu]
    })
    
    return cm_results


```

```python
# İlk modelimizden elde ettiğimiz hata matrisi
cm1 = conf_matrix(y_test, y_pred_cv1)
```

```python
# İkinci modelimizden elde ettiğimiz hata matrisi
cm2 = conf_matrix(y_test, y_pred_cv2)
```

```python
# DataFrame'leri birleştirerek sonuçları oluşturma
results = pd.DataFrame({
    'Metric': cm1['Metric'],
    'LR(Word)-CV': cm1['Score'],
    'LR(N-Grams)-CV': cm2['Score']
})

# İndeks ayarlama
results.set_index('Metric', inplace=True)

print(results)
```

##Naive Bayes algoritması metin verileri üzerinden geliştirilen modellerde genellikle oldukça başarılı çalışan bir algoritmadır. O zaman bir de Naive Bayes'den elde ettiğimiz sonuçlara bakalım.

```python
# İlk olarak MultinomialNB algoritmasıyla modellerimizi eğitelim
from sklearn.naive_bayes import MultinomialNB

mnb1 = MultinomialNB()
mnb1.fit(X_train_cv1, y_train)

mnb2 = MultinomialNB()
mnb2.fit(X_train_cv2, y_train)

y_pred_cv1_mnb = mnb1.predict(X_test_cv1)
y_pred_cv2_mnb = mnb2.predict(X_test_cv2)
```

```python
# Ardından da BernoulliNB algoritmasıyla
from sklearn.naive_bayes import BernoulliNB

bnb1 = BernoulliNB()
bnb1.fit(X_train_cv1, y_train)

bnb2 = BernoulliNB()
bnb2.fit(X_train_cv2, y_train)

y_pred_cv1_bnb = bnb1.predict(X_test_cv1)
y_pred_cv2_bnb = bnb2.predict(X_test_cv2)
```

```python
# MultinomialNB & Kelime
cm3 = conf_matrix(y_test, y_pred_cv1_mnb)
```

```python
# MultinomialNB & 2'li N-Grams
cm4 = conf_matrix(y_test, y_pred_cv2_mnb)
```

```python
# BernoulliNB & Kelime
cm5 = conf_matrix(y_test, y_pred_cv1_bnb)
```

```python
# BernoulliNB & 2'li N-Grams
cm6 = conf_matrix(y_test, y_pred_cv2_bnb)
```

```python
# Sonuçları DataFrame'e dönüştürme
results_nb = pd.DataFrame({
    'Metric': cm3['Metric'],
    'MNB(Word)-CV': cm3['Score'],
    'MNB(N-Grams)-CV': cm4['Score'],
    'BNB(Word)-CV': cm5['Score'],
    'BNB(N-Grams)-CV': cm6['Score']
})

# İndeksi ayarlama
results_nb.set_index('Metric', inplace=True)

# Sonuçları birleştirme
results = pd.concat([results, results_nb], axis=1)
print(results)
```

```python
import joblib
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression

# Veri hazırlama
df = pd.read_csv("sentiment_data.csv")
df= df.dropna()
X = df['temizlenmiş_yorum']
y = df['label']

# Veriyi eğitim ve test olarak ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# CountVectorizer ile kelimeleri sayma
cv1 = CountVectorizer(max_df=0.8, max_features=10000)
X_train_cv1 = cv1.fit_transform(X_train)
X_test_cv1 = cv1.transform(X_test)

# Logistic Regression modeli oluşturma ve eğitme
lr = LogisticRegression()
lr.fit(X_train_cv1, y_train)

# Model ve CountVectorizer'ı birlikte kaydetme
joblib.dump((lr, cv1), 'logistic_regression_model_with_vectorizer.pkl')
```
